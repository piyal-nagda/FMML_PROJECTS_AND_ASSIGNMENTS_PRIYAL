### Gradient Descent

````python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)  # Compute gradient using the derivative
    new_x = previous_x - step_multiplier * gradient

    step_size = abs(new_x - previous_x)
    # print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
````

## Gradient Descent

````python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Define the function and its derivative
def h(x):
    return x**5 - 2*x**4 + 2  # Given function

def dh(x):
    return 5*x**4 - 8*x**3  # Derivative of the function

# Gradient descent implementation
def gradient_descent_v3(derivative_func, initial_guess, multiplier=0.01, precision=0.0001, max_iter=1000):
    x = initial_guess
    steps = [x]

    for _ in range(max_iter):
        grad = derivative_func(x)
        new_x = x - multiplier * grad  # Gradient descent update
        steps.append(new_x)

        # Check convergence
        if abs(new_x - x) < precision:
            break

        x = new_x

    return steps

# Run gradient descent
initial_guess = 1.5  # Example initial guess
steps = gradient_descent_v3(dh, initial_guess, multiplier=0.01)

# Plot the function and gradient descent steps
x_vals = np.linspace(-1, 2.5, 500)  # Adjust the range for better visualization
y_vals = h(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="h(x) = x^5 - 2x^4 + 2", color="blue")
plt.scatter(steps, [h(x) for x in steps], color="red", label="Gradient Descent Steps")
plt.title("Function h(x) with Gradient Descent Steps")
plt.xlabel("x")
plt.ylabel("h(x)")
plt.legend()
plt.show()
````

````python
## TASK-7 : Plot reduction in cost for each iteration

import matplotlib.pyplot as plt
import numpy as np

# Define the function and its derivative
def h(x):
    return x**5 - 2*x**4 + 2  # Given function

def dh(x):
    return 5*x**4 - 8*x**3  # Derivative of the function

# Gradient descent with cost tracking
def gradient_descent_v4(derivative_func, initial_guess, multiplier=0.01, precision=0.0001, max_iter=1000):
    x = initial_guess
    steps = [x]
    cost_history = [h(x)]  # Track cost (function value) at each step

    for _ in range(max_iter):
        grad = derivative_func(x)
        new_x = x - multiplier * grad  # Gradient descent update
        steps.append(new_x)
        cost_history.append(h(new_x))  # Track cost

        # Check convergence
        if abs(new_x - x) < precision:
            break

        x = new_x

    return steps, cost_history

# Run gradient descent and track cost
initial_guess = 1.5  # Example initial guess
steps, cost_history = gradient_descent_v4(dh, initial_guess, multiplier=0.01)

# Plot the cost reduction
plt.figure(figsize=(10, 6))
plt.plot(cost_history, label="Cost (h(x)) at each iteration", color="green")
plt.title("Cost Reduction with Gradient Descent")
plt.xlabel("Iterations")
plt.ylabel("Cost (h(x))")
plt.legend()
plt.show()
````

### Example 4 - Data Viz with 3D Charts

#### Minimise $$f(x, y) = \frac{1}{3^{-x^2 - y^2} + 1}$$

Minimise $$f(x, y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$

````python
## TASK-8 : Code the same things as they were done for the first two examples
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function f(x, y) = 1 / (r + 1) where r = 3^(-x^2 - y^2)
def f(x, y):
    r = 3**(-x**2 - y**2)
    return 1 / (r + 1)

# Derivative of f(x, y) with respect to x and y
def df(x, y):
    r = 3**(-x**2 - y**2)
    df_dx = (-2 * x * r * np.log(3)) / (r + 1)**2
    df_dy = (-2 * y * r * np.log(3)) / (r + 1)**2
    return df_dx, df_dy

# Gradient descent to minimize the function
def gradient_descent_2d(derivative_func, initial_guess, multiplier=0.01, precision=0.0001, max_iter=1000):
    x, y = initial_guess
    steps = [(x, y)]

    for _ in range(max_iter):
        grad_x, grad_y = derivative_func(x, y)
        x_new = x - multiplier * grad_x
        y_new = y - multiplier * grad_y

        # Track the steps taken
        steps.append((x_new, y_new))

        # Check for convergence
        if np.sqrt((x_new - x)**2 + (y_new - y)**2) < precision:
            break

        x, y = x_new, y_new

    return steps

# Run gradient descent starting from an initial guess
initial_guess = (0.5, 0.5)  # Example initial guess
steps = gradient_descent_2d(df, initial_guess, multiplier=0.1)

# Create a meshgrid for plotting the function
x_vals = np.linspace(-2, 2, 100)
y_vals = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

# Plot the function in 3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Plot the gradient descent steps
steps_x, steps_y = zip(*steps)
steps_z = [f(x, y) for x, y in steps]
ax.plot(steps_x, steps_y, steps_z, color='r', marker='o', markersize=5, label='Gradient Descent Path')

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y)')
ax.set_title('3D Surface Plot of f(x, y) and Gradient Descent Path')
ax.legend()

plt.show()
````

#### Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

#### $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

#### $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

#### MSE & Gradient Descent

````python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
````

````python
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(y_5, thetas[0] + thetas[1]*x_5))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5))
````
